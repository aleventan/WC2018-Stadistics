```python
from game_def import openJsonMatches, getIdMatch, teamList
from plot_graph import plotGraph
from team_def import counterStats, separeteTeams
#from match import match
import glob
import os
from circular_graph import pieGraph
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
# Importing the airium library
from airium import Airium
#
# class generate_HTML():
#     global selected_team


def gen_HTML(selected_team):
    #from tasks import refreshed_google_client
    a = Airium()
    # Generating HTML file
    a('<!DOCTYPE html>')
    with a.html(lang="pl"):
        with a.head():
            a.meta(charset="utf-8")
            a.title(_t="Stadistics World Cup 2018")
            a.link(href="{{ url_for(\'static\',filename=\'styles/style.css\') }}", rel='stylesheet')
            a.link(href="{{ url_for(\'static\',filename=\'js/js.js\') }}", type='text/javascript')
        with a.body():
            matches = getIdMatch(selected_team)
            # Arma una nueva lista con los datos de cada partido por separado
            res = list(zip(*matches))
            print(res)
            cs = counterStats(selected_team)
            lt = teamList()
            with a.p(klass='p_combo1'):
                with a.div():
                    with a.form(method="post", action='/'):
                        with a.select(name="combo_team", klass= "combo1"):
                            with a.option(selected=True):
                                a("Choose a team")
                            for l in lt:
                                # if lt.index(l) % 4:
                                #     with a.optgroup():
                                #         a("Group A")
                                with a.option(value=l):
                                    a(l)
                        a.input(type="submit", value="Enviar")
                a('<br><br><br>')
            for c in cs:
                ht = c.get('home_team_name')
                at = c.get('away_team_name')
                hp = c.get('count_poss_home')
                ap = c.get('count_poss_away')
                with a.div(klass="div_left"):
                    with a.table():
                        with a.tr():
                            with a.td(klass="result_left"):
                                for r in res:
                                    if cs.index(c) == res.index(r):
                                        if c.get('count_penalties_home') > 0 and c.get('count_penalties_away') > 0:
                                            a('(' + str(c.get('count_penalties_home')) + ')')
                                        a(r[5])
                                flag_home_team = "{{ url_for(\'static\',filename=\'./flags/\') }}" + ht + ".png"
                                a.img(src=flag_home_team, alt=ht, klass="flags", align="right")
                            with a.td(klass = "date"):
                                for r in res:
                                    if cs.index(c) == res.index(r):
                                        date_m = str(r[1].split('-'))
                                        date_match = date_m[16:18] + '-' + date_m[10:12] + '-' + date_m[2:6] + '<br>'
                                        a(date_match)
                                        a(r[2])
                            with a.td():
                                for r in res:
                                    if cs.index(c) == res.index(r):
                                        a(r[6])
                                        if c.get('count_penalties_home') > 0 and c.get('count_penalties_away') > 0:
                                            a('(' + str(c.get('count_penalties_away')) + ')')
                                flag_away_team_name = "{{ url_for(\'static\',filename=\'./flags/\') }}" + at + ".png"
                                a.img(src=flag_away_team_name, alt=at, klass="flags", align="left")
                        with a.tr():
                            with a.th(klass = "leftside"):
                                a(ht)
                            with a.th():
                                a("Stadistics")
                            with a.th(klass = "rightside"):
                                a(at)
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(str(hp)+ '%')
                            with a.td(klass = "centerside"):
                                a("Possesion")
                            with a.td(klass = "rightside"):
                                a(str(ap) + '%')
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(c.get('count_shot_home'))
                            with a.td(klass = "centerside"):
                                a("Shots")
                            with a.td(klass = "rightside"):
                                a(c.get('count_shot_away'))
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(c.get('count_ball_recovery_home'))
                            with a.td(klass = "centerside"):
                                a("Ball Recovery")
                            with a.td(klass = "rightside"):
                                a(c.get('count_ball_recovery_away'))
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(c.get('count_fouls_committed_home'))
                            with a.td(klass = "centerside"):
                                a("Fouls Committed")
                            with a.td(klass = "rightside"):
                                a(c.get('count_fouls_committed_away'))
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(c.get('count_corner_home'))
                            with a.td(klass = "centerside"):
                                a("Corners")
                            with a.td(klass = "rightside"):
                                a(c.get('count_corner_away'))
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(c.get('count_block_home'))
                            with a.td(klass = "centerside"):
                                a("Blocks")
                            with a.td(klass = "rightside"):
                                a(c.get('count_block_away'))
                        with a.tr():
                            with a.td(klass = "leftside"):
                                a(c.get('count_interception_home'))
                            with a.td(klass = "centerside"):
                                a("Interception")
                            with a.td(klass = "rightside"):
                                a(c.get('count_interception_away'))
                with a.div(klass="div_center"):
                    name_file_circle = './static/possgraph/' + ht + '-' + at + '_stats.png'
                    file_circle_graph = "{{ url_for(\'static\',filename=\'./possgraph/\') }}" + ht + '-' + at + "_poss.png"
                    if os.path.exists(name_file_circle):
                        a.img(src=file_circle_graph, alt=file_circle_graph, klass="circlegraph", align="right")
                    else:
                        new_circle = pieGraph(ht, at, hp, ap)
                        new_circle.circleG()
                        a.img(src=file_circle_graph, alt=file_circle_graph, klass="circlegraph", align="right")
                with a.div(klass="div_right"):
                    st = separeteTeams(selected_team)
                    name_file_bar = './static/team_stats/' + ht + '-' + at + '_stats.png'
                    file_bars_graph = "{{ url_for(\'static\',filename=\'./team_stats/\') }}" + ht + '-' + at + "_stats.png"
                    for s in st[0]:
                        if cs.index(c) == st[0].index(s):
                            primero = s
                    for t in st[1]:
                        if cs.index(c) == st[1].index(t):
                            segundo = t
                    if os.path.exists(name_file_bar):
                        a.img(src=file_bars_graph, alt=file_bars_graph, klass="baregraph", align="right")
                    else:
                        new_plot = plotGraph(primero, segundo)
                        new_plot.plotG()
                        a.img(src=file_bars_graph, alt=file_bars_graph, klass="bargraph", align="right")






    # Casting the file to a string to extract the value
    html = str(a)
    # Casting the file to UTF-8 encoded bytes:
    html_bytes = bytes(a)
    print(html)
    h = open('templates/index.html', 'w')
    h.write(str(html))

    # close the file
    h.close()
```
